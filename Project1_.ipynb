```python
#1
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#2
df1=pd.read_csv("nhanes_adult_female_bmx_2020.csv")
df1.head()
```

```python
df2=pd.read_csv("nhanes_adult_male_bmx_2020.csv")
df2.head()
```

```python
#3 On a single plot (use matplotlib.pytplot.subplot), draw two histograms: for female
# weights (top subfigure). and for male weights (bottom subfigure) Call
# matplotlib.pyplot.xlim to make the xaxis limits identical for both subfigures (work out the
# appropriate limits yourself)

# Now we have to calculate the min and max weights for both datasets
min_weight = min(df1['BMXWT'].min(), df2['BMXWT'].min())
max_weight = max(df1['BMXWT'].max(), df2['BMXWT'].max())

plt.figure(figsize=(10,6))
# This is for female weights
plt.subplot(2, 1, 1)
plt.hist(df1['BMXWT'], bins=20, color='skyblue', edgecolor='black')
plt.title('Female Weights')
plt.ylabel('Frequency')
plt.xlim(min_weight, max_weight)

# this is for male weights
plt.subplot(2, 1, 2)
plt.hist(df2['BMXWT'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Male Weights')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.xlim(min_weight, max_weight)

plt.plot()
plt.show()
```

```python
# 4 Call matplotlib.pyplot.boxplot to draw a box-and-whisker plot, with two boxes side by
# side, giving the male and female weights so that they can be compared to each other.
# Note that the boxplot function can be fed with a list of two vectors like [female_weights,
# male_weights]. In your own words, discuss the results.


#drop the null and unknown values
female_weights = df1['BMXWT'].dropna()
male_weights = df2['BMXWT'].dropna()

plt.figure(figsize=(8, 6))
plt.boxplot([female_weights, male_weights], labels=['Female', 'Male'])
plt.title('Comparison of Female and Male Weights')
plt.ylabel('Weight (kg)')
plt.show()

# Discussion:
# Using this box plots we can see that differences between both gender's weights.
# the commoon thing is that we have a lot of outliers in both genders which different from normal set of data.
# the plotting is done by putting weights in yaxis and genders across in xaxis
# the male gender box plot's initial point is higher than the female's box plot
```

```python
#5 Compute the basic numerical aggregates of the male and female weights (measures
# of location, dispersion, and shape). In your own words, describe and compare the two
# distributions (e.g., are they left skewed, which one has more dispersion, and so forth).

#female weights
female_weight_stats = df1['BMXWT'].describe()
female_median = df1['BMXWT'].median()
female_skew = df1['BMXWT'].skew()

#male weights
male_weight_stats = df2['BMXWT'].describe()
male_median = df2['BMXWT'].median()
male_skew = df2['BMXWT'].skew()

print("Female Weight Stats:")
print(female_weight_stats)
print(f"Median: {female_median}")
print(f"Skewness: {female_skew}")

print("\nMale Weight Stats:")
print(male_weight_stats)
print(f"Median: {male_median}")
print(f"Skewness: {male_skew}")

# description and comparisons:
# The mean and median weights for males are generally higher than those for females,
# indicating that males tend to weigh more on average..
# A positive skew suggests a longer right tail, while a negative skew indicates a longer left tail.
# If the skewness values differ substantially, it means one distribution is more skewed than the other.
# Visually comparing the histograms and box plots provides further insight into the shapes of the distributions.
```

```python
#6 To the female matrix, add the eight column which gives the body mass indices of all
# the female participants

# Calculate BMI (Body Mass Index)
df1['BMI'] = df1['BMXWT'] / ((df1['BMXHT'] / 100) ** 2)

# Display the updated DataFrame with the BMI column
print(df1.head())
```

```python
#7 Create a new matrix zfemale being a version of the female dataset with all its columns
# standardised (by computing the z-scores of each column).

zfemale = (df1 - df1.mean()) / df1.std()
print(zfemale.head())
```

```python
#8 Draw a scatterplot matrix (pairplot) for the standardised versions of height, weight,
# waist circumference, hip circumference, and BMI of the females (based on zfemale).
# Compute Pearsons and Spearmans correlation coefficients for all pairs of variables.
# Interpret the obtained results.

import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant columns for the pairplot
cols_for_pairplot = ['BMXHT', 'BMXWT', 'BMXWAIST', 'BMXHIP', 'BMI']
zfemale_subset = zfemale[cols_for_pairplot]

# Create the pairplot
sns.pairplot(zfemale_subset)
plt.show()

# Calculate Pearson's correlation
pearson_corr = zfemale_subset.corr(method='pearson')
print("Pearson's Correlation Coefficients:\n", pearson_corr)

# Calculate Spearman's correlation
spearman_corr = zfemale_subset.corr(method='spearman')
print("\nSpearman's Correlation Coefficients:\n", spearman_corr)

# Observations are:
# Examine the pairplot for visual relationships between variables.
# Look for strong positive or negative correlations in the scatterplots.
# The Pearson's correlation coefficients quantify linear relationships,
# while Spearman's correlation assesses monotonic relationships.
# Compare the values in the correlation matrices.  High absolute values (close to 1 or -1)
# suggest strong relationships, while values near 0 indicate weak or no linear/monotonic relationships.
# For example, a strong positive correlation between weight and BMI is expected.
# Discuss any surprising or unexpected correlations observed.
```

```python
#9 Compute the waist circumference to height ratio and the waist circumference to hip
# circumference ratio of the male and female participants by adding two more columns to
# the males and females matrices

# Female
# Waist-to-Height Ratio
df1['Waist_to_Height_Ratio'] = df1['BMXWAIST'] / df1['BMXHT']

# Waist-to-Hip Ratio
df1['Waist_to_Hip_Ratio'] = df1['BMXWAIST'] / df1['BMXHIP']

# Male
# Calculate Waist-to-Height Ratio
df2['Waist_to_Height_Ratio'] = df2['BMXWAIST'] / df2['BMXHT']

# Calculate Waist-to-Hip Ratio
df2['Waist_to_Hip_Ratio'] = df2['BMXWAIST'] / df2['BMXHIP']

print(df1.head())
print(df2.head())
```

```python
#10 Draw a box-and-whisker plot with four boxes side by side, comparing the distribution
# of the waist to-height ratio and the waist-to-hip ratio of both male and female participants.
# Explain what you see.

# Combine the data for plotting
data_to_plot = [df1['Waist_to_Height_Ratio'].dropna(),
                df1['Waist_to_Hip_Ratio'].dropna(),
                df2['Waist_to_Height_Ratio'].dropna(),
                df2['Waist_to_Hip_Ratio'].dropna()]

labels = ['Female Waist-to-Height', 'Female Waist-to-Hip',
          'Male Waist-to-Height', 'Male Waist-to-Hip']

plt.figure(figsize=(10, 6))
plt.boxplot(data_to_plot, labels=labels)
plt.title('Comparison of Waist Ratios (Male vs. Female)')
plt.ylabel('Ratio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Explanation of the Box-and-Whisker Plot

# This box-and-whisker plot compares the distributions of waist-to-height and waist-to-hip
# ratios for males and females.
# Each box represents the interquartile range (IQR)  the middle 50% of the data  with the line
# inside the box showing the median.
# The whiskers extend to the most extreme data points within 1.5 times the IQR from the
# quartiles.  Points beyond the whiskers are
# considered outliers and are plotted individually.
# By comparing the boxes and whiskers, we can observe differences in the central tendency,
# spread, and presence of outliers for each group.
```

11. BMI
Advantages:
Easy to calculate: Only requires weight and height.
Widely used for health and physical analysis
Baseline health indicator can be identified

Disadvantages:
Ignores body composition
No fat distribution info: Fails to show where fat is stored, which is important for health risks.
Population bias, as it is not suitable for all population of a country or region

2. Waist-to-Height Ratio

Advantages:
Accounts for fat distribution
Better risk predictor
Universal cutoff: A simple rule: your waist should be less than half your height.

Disadvantages:
Less common: Not as well-known or standardized as BMI, so less often used in clinical or population studies.
Measurement variability: Waist circumference can vary slightly depending on measurement technique (e.g., relaxed or contracted stomach).

3. Waist-to-Hip Ratio

Advantages:
Focus on fat distribution, focussed on obesity
Better for certain populations, used for population with diversity

Disadvantages:
Measurement challenges
Not as intuitive, not easy to understand for all.
